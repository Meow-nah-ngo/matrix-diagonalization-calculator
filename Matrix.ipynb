# Matrix format

```python
def print_matrix(matrix, name="Matrix"):

    print(f"\n{name}:")

    for row in matrix:
        print("[ ", end="")

        for value in row:
            print(f"{value:10.3f}", end=" ")

        print("]")
```

## Input

```python
def input_matrix():

    while True:

        size = int(input("Enter matrix size (2 or 3): "))

        if size == 2 or size == 3:
            break

        print("Invalid size. Please enter 2 or 3.")

    matrix = []

    print(f"\nEnter {size}x{size} matrix:")

    for i in range(size):

        while True:

            row = list(map(float, input(f"Row {i+1}: ").split()))

            if len(row) == size:
                matrix.append(row)
                break

            print(f"Please enter exactly {size} numbers.")

    return matrix
```

# DETERMINANT

## 2x2

```python
def det2x2(m):

    return m[0][0]*m[1][1] - m[0][1]*m[1][0]
```

## 3x3

```python
def det3x3(m):

    a,b,c = m[0]
    d,e,f = m[1]
    g,h,i = m[2]

    return (
        a*(e*i - f*h)
        - b*(d*i - f*g)
        + c*(d*h - e*g)
    )
```

# SUBTRACT λI

```python
def subtract_lambda_I(matrix, lamb):

    result = []

    for i in range(3):

        row = []

        for j in range(3):

            if i == j:
                row.append(matrix[i][j] - lamb)

            else:
                row.append(matrix[i][j])

        result.append(row)

    return result
```

# EIGENVALUES 

## 2x2

```python
def eigenvalues_2x2(matrix):

    a = matrix[0][0]
    b = matrix[0][1]
    c = matrix[1][0]
    d = matrix[1][1]

    trace = a + d
    determinant = det2x2(matrix)

    discriminant = trace**2 - 4*determinant

    if discriminant < 0:
        return []

    l1 = (trace + math.sqrt(discriminant)) / 2
    l2 = (trace - math.sqrt(discriminant)) / 2

    return [l1, l2]
```

## 3x3

```python
# (Triangular Matrix Only)
def is_triangular(matrix):

    size = len(matrix)

    upper = True
    lower = True

    for i in range(size):
        for j in range(size):

            if i > j and matrix[i][j] != 0:
                upper = False

            if i < j and matrix[i][j] != 0:
                lower = False

    return upper or lower
```

```python
def eigenvalues_3x3(matrix):

    if is_triangular(matrix):

        return [
            matrix[0][0],
            matrix[1][1],
            matrix[2][2]
        ]

    else:
        return None
```

```python
def eigenvalues_3x3(matrix):

    eigenvalues = []

    for lamb in range(-10, 11):

        test = subtract_lambda_I(matrix, lamb)

        determinant = det3x3(test)

        if abs(determinant) < 0.0001:

            eigenvalues.append(lamb)

    return eigenvalues
```

# EIGENVECTOR

## 2x2

```python
def eigenvector_2x2(matrix, eigenvalue):

    a = matrix[0][0] - eigenvalue
    b = matrix[0][1]

    c = matrix[1][0]
    d = matrix[1][1] - eigenvalue

    if b != 0:
        return [1, -a/b]

    elif c != 0:
        return [-d/c, 1]

    else:
        return [1, 0]
```

## 3x3

```python
def gaussian_elimination(matrix):

    m = [row[:] for row in matrix]

    rows = len(m)
    cols = len(m[0])

    lead = 0

    for r in range(rows):

        if lead >= cols:
            return m

        i = r

        while m[i][lead] == 0:

            i += 1

            if i == rows:

                i = r
                lead += 1

                if cols == lead:
                    return m

        m[i], m[r] = m[r], m[i]

        lv = m[r][lead]

        m[r] = [x / lv for x in m[r]]

        for i in range(rows):

            if i != r:

                lv = m[i][lead]

                m[i] = [
                    iv - lv*rv
                    for rv, iv in zip(m[r], m[i])
                ]

        lead += 1

    return m
```

```python
def eigenvector_3x3(matrix, eigenvalue):

    system = subtract_lambda_I(matrix, eigenvalue)

    reduced = gaussian_elimination(system)

    print("\nReduced Matrix:")

    for row in reduced:
        print(row)

    print("\nApproximate Eigenvector:")

    # free variable z = 1

    z = 1

    y = 0
    x = 0

    if reduced[0][0] != 0:

        x = -reduced[0][2] / reduced[0][0]

    if reduced[1][1] != 0:

        y = -reduced[1][2] / reduced[1][1]

    return [x, y, z]
```

# CREATE P MATRIX

```python
def create_P(vectors):

    size = len(vectors)

    P = []

    for i in range(size):

        row = []

        for vec in vectors:
            row.append(vec[i])

        P.append(row)

    return P
```

# CREATE D MATRIX

```python
def create_D(eigenvalues):

    size = len(eigenvalues)

    D = []

    for i in range(size):

        row = []

        for j in range(size):

            if i == j:
                row.append(eigenvalues[i])
            else:
                row.append(0)

        D.append(row)

    return D
```

# INVERSE 2x2

```python
def inverse_2x2(matrix):

    det = det2x2(matrix)

    if det == 0:
        return None

    a = matrix[0][0]
    b = matrix[0][1]
    c = matrix[1][0]
    d = matrix[1][1]

    return [
        [d/det, -b/det],
        [-c/det, a/det]
    ]
```

# CHECK DIAGONALIZABLE

```python
def is_diagonalizable_2x2(eigenvalues):

    unique = set(eigenvalues)

    return len(unique) == 2
```
